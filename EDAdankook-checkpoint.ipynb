```python
rmse = lambda y, p: np.sqrt(mean_squared_error(y, p))
```

```python
target_col = 'class'   # class를 target으로 한다.
pred_col = 'pred'   # 예측값을 저장하는 변수 선언
num_cols = ['g', 'dered_g'] #예측값 확인 
```

```python
lr = LinearRegression()   # 선형회귀 실행
lr.fit(df3[num_cols], df3[target_col])   # 선형회귀모델 학습
```

```python
df3[pred_col] = lr.predict(df3[num_cols]) # 선형회귀 후 예측값 저장
```

```python
df3.head() # 저장 후 결과 확인(pred)
```

```python
lr = LinearRegression()
lr.fit(df3[num_cols], df3[target_col])
df3[pred_col] = lr.predict(df3[num_cols])

print(f' RMSE:\t{rmse(df3[target_col], df3[pred_col]):12.2f}')
```
